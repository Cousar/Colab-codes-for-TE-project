```python
# ============================
#100MeV
# Mount Google Drive
# ============================
from google.colab import drive
drive.mount('/content/drive')

import os

# ==========================================
# Configure output directory in Google Drive
# ==========================================
# EXAMPLES:
# - Folder in "My Drive"  : "/content/drive/MyDrive/WPE_PLDs"
# - Folder in Shared drive: "/content/drive/Shared drives/<DriveName>/WPE_PLDs"
#
# Change this to the exact path of your (shared) folder:
OUTPUT_DIR = "/content/drive/Othercomputers/My Laptop/Germany/Germany main files/Essen/Ph.D/1.13-FA Cup Project/1.11-PLDfilesCreation/1.3-XShiftSeries/100MeV"  # <-- ADAPT THIS

os.makedirs(OUTPUT_DIR, exist_ok=True)

# ============================
# PLD writing helper function
# ============================
def write_single_spot_pld_xshift(x_mm, MU, energy=100.0, y=0.0, repainting=1):
    """
    Write a PLD file with a single spot at (x_mm, y) [IEC coordinates] into OUTPUT_DIR.

    Parameters
    ----------
    x_mm : float
        X position in IEC coordinate system in isocenter plane [mm].
    MU : float
        Total MU to be delivered on this single spot.
    energy : float
        Beam energy at nozzle exit in MeV.
    y : float
        Y position in IEC coordinate system in isocenter plane [mm].
    repainting : int
        Number of repaintings (does not change total delivered MU).
    """

    cumulative_msw = 1          # sum of meterset weights of all spots
    number_of_layers = 1
    number_of_elements = 2      # two elements per spot
    total_MU = 100               # beam meterset (total MU)

    content = (
        "Beam,Patient ID,Patient Name,Patient Initial,Patient Firstname,"
        f"Plan Label,Beam Name,{total_MU},{cumulative_msw},{number_of_layers}\n"
        f"Layer,4.0,{energy},{cumulative_msw},{number_of_elements},{repainting}\n"
        f"Element,{x_mm},{y},0,0\n"
        f"Element,{x_mm},{y},1,0\n"
    )

    # Build a nice filename, e.g. single_spot_100MeV_X+01mm.pld
    sign = "p" if x_mm >= 0 else "m"
    x_abs = abs(int(x_mm))
    base_name = f"single_spot_100MeV_X{sign}{x_abs:02d}mm"

    pld_path = os.path.join(OUTPUT_DIR, base_name + ".pld")
    with open(pld_path, "w", encoding="utf-8") as f:
        f.write(content)

    print(f"Created: {pld_path}")

# ============================
# Main: generate all X shifts
# ============================
if __name__ == "__main__":
    energy = 100.0
    y_pos = 0.0          # Y stays at 0 mm (IEC)
    repainting = 1

    # Set MU once for all spots (adapt as needed)
    MU = 100.0

    # X shifts from -22 mm to +22 mm in 1 mm steps (including 0)
    x_positions = range(-22, 23, 1)

    for x in x_positions:
        write_single_spot_pld_xshift(x_mm=float(x), MU=MU, energy=energy, y=y_pos, repainting=repainting)
```
