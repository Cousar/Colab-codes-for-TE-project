```python
# ============================
# Mount Google Drive
# ============================
from google.colab import drive
drive.mount('/content/drive')

import os

# ==========================================
# Configure output directory in Google Drive
# ==========================================
# EXAMPLES:
# - Folder in "My Drive"  : "/content/drive/MyDrive/WPE_PLDs"
# - Folder in Shared drive: "/content/drive/Shared drives/<DriveName>/WPE_PLDs"
#
# Change this to the exact path of your (shared) folder:
OUTPUT_DIR = "/content/drive/Othercomputers/My Laptop/Germany/Germany main files/Essen/Ph.D/1.13-FA Cup Project/PLDfilesCreation/SingleSpotsDiffMU"  # <-- ADAPT THIS

os.makedirs(OUTPUT_DIR, exist_ok=True)

# ============================
# PLD writing helper function
# ============================
def write_single_spot_pld(base_name, MU, energy=100.0, x=0.0, y=0.0, repainting=1):
    """
    Write a PLD file with a single spot at (x, y) into OUTPUT_DIR.

    Parameters
    ----------
    base_name : str
        Base name without extension (e.g. 'single_spot_100MeV_1MU').
    MU : float
        Total MU to be delivered on this single spot.
    energy : float
        Beam energy at nozzle exit in MeV.
    x, y : float
        Spot position in IEC coordinates in isocenter plane [mm].
    repainting : int
        Number of repaintings (does not change total delivered MU).
    """

    cumulative_msw = 1          # sum of meterset weights of all spots
    number_of_layers = 1
    number_of_elements = 2      # two elements per spot
    total_MU = MU               # beam meterset (total MU)

    content = (
        "Beam,Patient ID,Patient Name,Patient Initial,Patient Firstname,"
        f"Plan Label,Beam Name,{total_MU},{cumulative_msw},{number_of_layers}\n"
        f"Layer,4.0,{energy},{cumulative_msw},{number_of_elements},{repainting}\n"
        f"Element,{x},{y},0,0\n"
        f"Element,{x},{y},1,0\n"
    )

    # Write .pld for TCS into the Drive folder
    pld_path = os.path.join(OUTPUT_DIR, base_name + ".pld")
    with open(pld_path, "w", encoding="utf-8") as f:
        f.write(content)

    print(f"Created: {pld_path}")

# ============================
# Main: generate all MU files
# ============================
if __name__ == "__main__":
    mu_values = [
        0.025,
        0.05,
        0.1,
        0.2,
        0.4,
        1,
        2,
        5,
        10,
        25,
        50,
        75,
        100,
        125,
        150,
        175,
        200,
        1000,
    ]

    energy = 100.0
    x_pos = 0.0
    y_pos = 0.0
    repainting = 1

    for mu in mu_values:
        mu_str = str(mu).replace(".", "p")  # e.g. 0.025 -> "0p025"
        base_name = f"single_spot_100MeV_{mu_str}MU"
        write_single_spot_pld(base_name, mu, energy=energy, x=x_pos, y=y_pos, repainting=repainting)
```
