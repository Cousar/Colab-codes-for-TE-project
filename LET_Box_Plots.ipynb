```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import ticker # Import the ticker module

# --- 1. Define Sample Data Structure ---
# Replace this with your actual data.
# Each inner dictionary should contain 'min', 'max', 'median', 'mean' for each box type.
# The values are in kilo electron volts per micrometer.
# Mean LETd distribution
data = {
    "Brain stem": {
        "CAP": {'median': 3.2, 'mean': 3.8 ,'min': 2.7, 'max': 5.9 },
        "DOSEopt": {'median': 3.1, 'mean': 3.7 ,'min': 2.6, 'max': 5.9 },
        "DOSE+LETopt (W)": {'median': 3.1, 'mean': 3.5 ,'min': 2.4, 'max': 5.5 },
        "DOSE+LETopt (C)": {'median': 3, 'mean': 3.4 ,'min': 2.4, 'max': 5.3 },
        "LETopt_LW": {'median': 3.1, 'mean': 3.4 ,'min': 2.4, 'max': 5.1 },
        "LETopt_HW": {'median': 3.1, 'mean': 3.3 ,'min': 2.3, 'max': 4.9 },
    },
    "Brain Stem core": {
      "CAP": {'median': 4, 'mean': 4.1 ,'min': 2.9, 'max': 6 },
        "DOSEopt": {'median': 3.8, 'mean': 4 ,'min': 2.8, 'max': 5.7},
        "DOSE+LETopt (W)": {'median': 3.8, 'mean': 3.7 ,'min': 2.4, 'max': 5.3 },
        "DOSE+LETopt (C)": {'median': 3.7, 'mean': 3.6 ,'min': 2.4, 'max': 5.2 },
        "LETopt_LW": {'median': 3.8, 'mean': 3.6 ,'min': 2.4, 'max': 4.9 },
        "LETopt_HW": {'median': 3.8, 'mean': 3.6 ,'min': 2.3, 'max': 4.9 },
    },
    "Optic nerves": {
   "CAP": {'median': 3 , 'mean': 3.2 ,'min': 2.2, 'max': 4.6 },
        "DOSEopt": {'median': 2.9, 'mean': 3.1 ,'min': 2.2, 'max': 4.8},
        "DOSE+LETopt (W)": {'median': 2.7, 'mean': 3 ,'min': 2.1, 'max': 4.9 },
        "DOSE+LETopt (C)": {'median': 2.9, 'mean': 3 ,'min': 2.1, 'max': 4.8 },
        "LETopt_LW": {'median': 2.7, 'mean': 2.9 ,'min': 2.2, 'max': 4.7 },
        "LETopt_HW": {'median': 2.6, 'mean': 2.8 ,'min': 2.1, 'max': 4.4 },
    },
    "Optic chiasm": {
   "CAP": {'median': 2.9, 'mean': 3.1 ,'min': 2.1, 'max': 4.5},
        "DOSEopt": {'median': 2.9, 'mean': 3.1 ,'min': 2.1, 'max': 4.6 },
        "DOSE+LETopt (W)": {'median': 2.4, 'mean': 2.8 ,'min': 2.1, 'max': 3.7 },
        "DOSE+LETopt (C)": {'median': 2.4, 'mean': 2.7 ,'min': 2.1, 'max': 3.6 },
        "LETopt_LW": {'median': 2.5, 'mean': 2.5 ,'min': 2.2, 'max': 2.9},
        "LETopt_HW": {'median': 2.4, 'mean': 2.4 ,'min': 2.1, 'max': 2.6 },
    },
    "Cochlea": {
  "CAP": {'median': 3.6, 'mean': 3.4 ,'min': 2.1, 'max': 4 },
        "DOSEopt": {'median': 3.5, 'mean': 3.2 ,'min': 2.1, 'max': 3.9 },
        "DOSE+LETopt (W)": {'median': 2.5, 'mean': 2.5 ,'min': 2.1, 'max': 2.9 },
        "DOSE+LETopt (C)": {'median': 2.5, 'mean': 2.5 ,'min': 2.1, 'max': 2.7 },
        "LETopt_LW": {'median': 2.5, 'mean': 2.5 ,'min': 2.1, 'max': 2.7 },
        "LETopt_HW": {'median': 2.4, 'mean': 2.4 ,'min': 2.2, 'max': 2.7 },
    },
    "PVR": {
   "CAP": {'median': 4.6, 'mean': 4.5 ,'min': 2.5, 'max': 6.3 },
        "DOSEopt": {'median': 4.2, 'mean': 4.5 ,'min': 2.4, 'max': 6.6 },
        "DOSE+LETopt (W)": {'median': 4.9, 'mean': 4.5 ,'min': 2.5, 'max': 6.1 },
        "DOSE+LETopt (C)": {'median': 4.3, 'mean': 4.4 ,'min': 2.5, 'max': 6.1},
        "LETopt_LW": {'median': 4.5, 'mean': 4.4 ,'min': 2.5, 'max': 6.1},
        "LETopt_HW": {'median': 4.4, 'mean': 4.3 ,'min': 2.4, 'max': 6.1 },
    },
}

# Define the order of OARs for plotting
oar_order = list(data.keys())

# Define the order and labels of box types
box_types_labels = {
    "CAP": "CAP",
    "DOSEopt": "DOSEopt",
    "DOSE+LETopt (W)": "DOSE+LETopt (W)",
    "DOSE+LETopt (C)": "DOSE+LETopt (C)",
    "LETopt_LW": "LETopt_LW",
    "LETopt_HW": "LETopt_HW",
}
box_types = list(box_types_labels.keys())

# --- Define colors for each box type using a brighter and cooler colormap ---
# FIX: Updated to suppress MatplotlibDeprecationWarning
colors = plt.colormaps.get_cmap('GnBu').resampled(len(box_types)) # 'GnBu' (Green-Blue) for brighter and cooler

# --- 2. Custom Plotting Function ---
def draw_custom_box(ax, x_pos, min_val, max_val, median_val, mean_val, box_width, box_color, label=None):
    """
    Draws a custom box plot showing range, median, and mean.

    Args:
        ax: The matplotlib axes to draw on.
        x_pos: The x-coordinate for the center of the box.
        min_val: The minimum value of the range.
        max_val: The maximum value of the range.
        median_val: The median value.
        mean_val: The mean value.
        box_width: The width of the box.
        box_color: The color for the box.
        label: The label for the legend (optional).
    """
    # Draw the main box (range)
    rect = plt.Rectangle((x_pos - box_width / 2, min_val), box_width, (max_val - min_val),
                         facecolor=box_color, edgecolor='black', linewidth=0.8, label=label)
    ax.add_patch(rect)

    # Draw the median line (now red)
    ax.plot([x_pos - box_width / 2, x_pos + box_width / 2], [median_val, median_val],
            color='red', linestyle='-', linewidth=2, zorder=3) # Changed to red, added zorder

    # Draw the mean marker (back to red diamond)
    ax.plot(x_pos, mean_val, marker='D', color='red', markersize=6, zorder=5) # Back to red diamond


# --- 3. Create the Plot ---
fig, ax = plt.subplots(figsize=(15, 8)) # Adjust figure size as needed

# Parameters for plotting
num_oars = len(oar_order)
num_box_types = len(box_types)
box_width = 0.08 # Narrow enough width for each individual box
spacing_within_oar = 0.1 # Spacing between box types within an OAR compartment

# Create x-tick positions for each OAR compartment
x_oar_positions = np.arange(num_oars)

# Iterate through each OAR
for i, oar_name in enumerate(oar_order):
    # Calculate the starting x-position for the first box within this OAR compartment
    start_x_for_oar = x_oar_positions[i] - ((num_box_types - 1) * spacing_within_oar + box_width) / 2

    # Iterate through each box type within the current OAR
    for j, box_type in enumerate(box_types):
        box_data = data[oar_name][box_type]
        current_x_pos = start_x_for_oar + j * spacing_within_oar

        # Get color for the current box type
        box_color = colors(j)

        # Draw the custom box plot
        # Only add label for the first OAR to avoid duplicate legend entries
        label = box_types_labels[box_type] if i == 0 else None
        draw_custom_box(ax, current_x_pos,
                        box_data['min'], box_data['max'],
                        box_data['median'], box_data['mean'],
                        box_width, box_color, label=label)

# --- 4. Labels, Title ---
ax.set_xticks(x_oar_positions)
ax.set_xticklabels(oar_order, rotation=45, ha='right') # Rotate for better readability
#ax.set_xlabel("Organ at Risk (OAR)")
ax.set_ylabel(r"LETd (keV / $\mu$m)") # Changed to use Mathtext for micro symbol
#ax.set_title("Mean LETd Distribution in Different Organs at Risk") # Added a more descriptive title
ax.set_title(" ") # Added a more descriptive title

# Set Y-axis tick spacing to 0.5
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.5))


# --- Legend for Box Types (Hue) ---
# We need to get the handles and labels from the patches added to the axes for the main legend.
# A small dummy rectangle can be used for each color to show in the legend.
legend_handles = []
for j, box_type_key in enumerate(box_types):
    legend_handles.append(plt.Rectangle((0, 0), 1, 1, fc=colors(j), ec='black', linewidth=0.8))
main_legend = ax.legend(handles=legend_handles, labels=[box_types_labels[bt] for bt in box_types],
                        bbox_to_anchor=(1.05, 1), loc='upper left',
                        fancybox=True, shadow=True, fontsize='small', title_fontsize='medium')

# --- Legend for Median and Mean ---
# Create dummy plots for legend entries. Use 'red' for both.
median_line_legend = plt.Line2D([0], [0], color='red', linestyle='-', linewidth=2)
mean_marker_legend = plt.Line2D([0], [0], marker='D', color='red', markersize=6, linestyle='None')

# Create the second legend directly below the first one
# We get the exact position of the first legend and calculate the new bbox_to_anchor
# based on its height.
renderer = fig.canvas.get_renderer()
main_legend_bbox = main_legend.get_window_extent(renderer=renderer)
main_legend_bbox_in_axes = main_legend_bbox.transformed(ax.transAxes.inverted())

# Calculate the new y position for the second legend, a bit below the first one
y_pos_second_legend = main_legend_bbox_in_axes.y0 - 0.05 # 0.05 is a small padding

second_legend = ax.legend(handles=[median_line_legend, mean_marker_legend],
                          labels=['Median', 'Mean'], # Labels are "Median" and "Mean" as requested
                          loc='upper left',
                          bbox_to_anchor=(main_legend_bbox_in_axes.x0, y_pos_second_legend),
                          fancybox=True, shadow=True, fontsize='small', title_fontsize='medium')

# Add both legends to the figure, not just the axes, to ensure they don't overwrite each other.
fig.add_artist(main_legend)
fig.add_artist(second_legend)


# --- 5. Adjust Layout ---
plt.grid(axis='y', linestyle='--', alpha=0.7) # Add horizontal grid lines
plt.tight_layout(rect=[0, 0, 0.85, 1]) # Adjust layout to make space for the legends
plt.show()
```
